```python
import pandas as pd 
import numpy as np 
import xlrd
import matplotlib.pyplot as plt
```

# Reading Excel file  

```python
df = pd.read_excel('sheetone.xls',sheetname ='Primary NAR',converters = {'Total':np.int32,
                                                                         'Gender(Male)':np.int32,
                                                                         'Gender(Female)':np.int32,
                                                                         'R(Urban)':np.int32,
                                                                         'R(Rural)':np.int32,
                                                                         'WQ(Poor)':np.int32,
                                                                         'WQ(Second)':np.int32,
                                                                         'WQ(Second)':np.int32,
                                                                         'WQ(Middle)':np.int32,
                                                                         'WQ(Fourth)':np.int32,
                                                                         'WQ(Richest)':np.int32,})
df
```

#  Removing NaN values 

```python

df1 = pd.DataFrame.dropna(df)
df1 = df1.reset_index(drop=True)
df1
```

```python
list(df1)
```

# Selecting required columns

```python
selected = df1[['ISO Code',
                'Countries and areas',
                'Reference year(s)',
                'WQ(Poor)',
                'WQ(Second)',
                'WQ(Middle)',
                'WQ(Fourth)',
                'WQ(Richest)']]
selected
```

# Selecting Asian countries

```python
asiac = ['AFG', 'ARM','AZE','BHR','BGD','BTN','BRN','KHM','CHN','CXR','CCK','IOT','GEO','HKG','IND','IDN','IRN','IRQ',
         'ISR','JPN','JOR','KAZ','PRK','KOR','KWT','KGZ','LAO','LBN','MAC','MYS','MDV','MNG','MMR','NPL','OMN','PAK','PHL',
         'QAT','SAU','SGP','LKA','SYR','TWN','TJK','THA','TUR','TKM','ARE','UZB','VNM','YEM','PSE']
         
```

```python
asian_countries = selected.loc[selected['ISO Code'].isin(asiac)]
asian_countries.reset_index(drop=True)
```

# selecting non Asian countries 

```python
non_asian_countries = selected.loc[~selected['ISO Code'].isin(asiac)]
non_asian_countries.reset_index(drop=True)
```
